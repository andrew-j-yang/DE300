```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_csv("On_Time_Reporting_Carrier_On_Time_Performance_2019_1.csv")
```

Variable Set #1: Airport Origin vs Departure Delay Time

This variable set attempts to demonstrate the average departure delay time for each airport. It is difficult to read the exact airport names due to the sheer number of them, but it is evident that there are more aiports that have large delay times than there are airports that have small/negative delay times. This demonstrates that you can typically expect flights to run late at most airports.

```python
average_delays = data.groupby('Origin')['DepDelay'].mean()

# Convert the Series to DataFrame for easier plotting
average_delays_df = average_delays.reset_index()

# Sort the values for better visualization
average_delays_df.sort_values(by='DepDelay', inplace=True)

# Create a bar plot
plt.figure(figsize=(10, 8))  # Set the figure size
plt.bar(average_delays_df['Origin'], average_delays_df['DepDelay'], color='blue')
plt.xlabel('Airport')  # Label on X-axis
plt.ylabel('Average Delay Time (minutes)')  # Label on Y-axis
plt.title('Average Delay Time by Airport')  # Title of the plot
plt.xticks(rotation=45)  # Rotate the X-axis labels for better readability
plt.tight_layout()  # Adjusts plot to ensure everything fits without overlap
plt.xticks(rotation=90)
plt.show()
```

Variable Set #2: Flight distance vs Flight air time

This variable set attempts to demonstrate the correlation between the distance of a flight, and the air time for a flight. We can observe that there are flights that are the same distance, but vary drastically in flight time. However, we can see the overall correlation demonstrates that the longer a flight is in terms of distance, the more flight time it will have.

```python
sns.lmplot(x='Distance', y='AirTime', data=data, aspect=1.5, scatter_kws={'alpha':0.5})
plt.title('Distance vs Flight Time')
plt.xlabel('Distance (miles)')
plt.ylabel('Flight Time (minutes)')
plt.grid(True)
plt.show()
```
